```python
import pandas as pd
import sqlite3

con = sqlite3.connect("./flyingv.db")
```

```python
c = con.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = c.fetchall()
res = []
for t in tables:
    res.append(t[0])
res
```

```python
dfs = []
for t in res:
    d = pd.read_sql_query("SELECT * from %s"%t, con)
    d['category'] = t
    dfs.append(d)
```

```python
df = pd.concat(dfs)
```

```python
df.to_excel('./flyingv.xlsx', engine='xlsxwriter')
```
